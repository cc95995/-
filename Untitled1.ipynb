```python
import pandas as pd
import numpy as np
data = pd.read_csv("C:/Users/zhangchi23/Desktop/10liushui.csv",encoding="gbk")
```

```python
data = pd.read_csv("C:/Users/zhangchi23/Desktop/11liushui.csv",encoding="gbk")
a = 'NULL'
data['区域'] = data['区域'].fillna(a)
data['业务线'] = data['业务线'].fillna(a)

s = data['业务线'].value_counts()
s1 = list(s.index)
s1.remove('外部合作业务')
data1 = data[data.业务线.isin(s1)]
data1['1级部门名称'][(data1['1级部门名称'] != '电脑数码事业部') & (data1['1级部门名称'] != '家电事业部') & (data1['1级部门名称'] != '通讯事业部') & (data1['1级部门名称'] != '居家生活事业部')] = '电脑数码事业部'
data1['区域'][data1['区域'] == '总部' ] = 'NULL'
# data1.to_csv("D:/pytest/design/data666.csv", index=False, encoding='gbk')
```

```python
#得到 业务线×部门 交叉表
tj1 = data1.ix[:,['业务线', '1级部门名称', '金额']]
tj11 = pd.pivot_table(tj1,index = ['业务线'], columns=['1级部门名称'], values=['金额'], aggfunc = np.sum)
tj11.columns = ['家电事业部', '居家生活事业部', '电脑数码事业部', '通讯事业部']
tj11 = tj11.fillna(0)
a = tj11.loc['EPBU',:] + tj11.loc['NULL',:]
tj11.loc['EPBU',:] = a
s2 = ['EPBU', 'AMBU', 'OPBU', 'SIBU']
tj11 = tj11[tj11.index.isin(s2)]
tj11['达成'] = tj11['家电事业部'] + tj11['居家生活事业部'] + tj11['电脑数码事业部'] +tj11['通讯事业部']
tj11.loc['合计',:] = tj11.loc['EPBU',:] + tj11.loc['AMBU',:] + tj11.loc['OPBU',:] + tj11.loc['SIBU',:]
```

```python
# 六大区域 × 品类 交叉表
tj2 = data1.ix[:,['区域', '1级部门名称', '金额']]
tj22 = pd.pivot_table(tj2,index = ['区域'], columns=['1级部门名称'], values=['金额'], aggfunc = np.sum)
tj22.columns = ['家电事业部', '居家生活事业部', '电脑数码事业部', '通讯事业部']
tj22 = tj22.fillna(0)
s3 = ['华北区', '华东区', '华南区', '西南区', '华中区', '东北区']
tj22 = tj22[tj22.index.isin(s3)]
tj22['达成'] = tj22['家电事业部'] + tj22['居家生活事业部'] + tj22['电脑数码事业部'] +tj22['通讯事业部']
tj22.loc['合计',:] = tj22.loc['华北区',:] + tj22.loc['华东区',:] + tj22.loc['华南区',:] + tj22.loc['西南区',:] + tj22.loc['华中区',:] + tj22.loc['东北区',:]
```

```python
#EPBU 的 各区域 × 品类 交叉表
data_epbu = data1[data1.业务线.isin(['EPBU', 'NULL'])]
tj33 = pd.pivot_table(data_epbu ,index = ['区域'], columns=['1级部门名称'], values=['金额'], aggfunc = np.sum)
tj33.columns = ['家电事业部', '居家生活事业部', '电脑数码事业部', '通讯事业部']
tj33 = tj33.fillna(0)
tj33.rename(index={'NULL':'其他', '宿迁':'电销'}, inplace=True)
tj33['达成'] = tj33['家电事业部'] + tj33['居家生活事业部'] + tj33['电脑数码事业部'] +tj33['通讯事业部']
tj33.loc['合计',:] = np.sum(tj33)
```
